### Task 5. Data Analysis|

### Univariate Analysis

```python
df.columns
```

```python
# Identifying the outliers in price column using boxplot
sns.boxplot(data = df, x = 'price')
plt.show()
```

```python
new_df = df.copy() # creating a copy of the original dataframe to perform outlier treatment.
```

```python
Q1 = new_df['price'].quantile(0.25)
Q3 = new_df['price'].quantile(0.75)
IQR = Q3 - Q1

print(f"{'price'} - Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"price - Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
new_df['price'].describe()
```

```python
df["price"].skew()
```

The IQR method identified an upper bound of 377 for the price column. However, the distribution of Airbnb prices is highly right-skewed, with a maximum value of 100,000 and a very large standard deviation (1023). This indicates the presence of extreme values that inflate the variance.

To avoid removing legitimate high-priced listings while still filtering unrealistic values, listings with prices greater than *1500* were treated as outliers and removed from the dataset.

```python
new_df = new_df[new_df["price"] <= 1500]
```

```python
# Boxplot after outlier treatment
sns.boxplot(data = new_df, x = 'price')
plt.show()
```

```python
new_df.shape
```

```python
# price distribution
plt.figure(figsize = (8,5))
sns.histplot(data = new_df, x = 'price',bins = 100)
plt.title("Price distribution")
plt.ylabel("Frequency")
plt.show()
```

### Bi Variable Analysis  
One variable dependancy on another variable

```python
room_price = new_df.groupby('room_type')['price'].mean().sort_values(ascending=False)
print(room_price)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=room_price.index, y=room_price.values)
plt.title("Average Price by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Average Price")
plt.xticks(rotation=30)
plt.show()
```

```python
host_counts = new_df['host_id'].value_counts()

multi_hosts = host_counts[host_counts > 1]
print("Number of hosts with multiple listings:", len(multi_hosts))
```

```python
top_hosts = host_counts.head(10)

plt.figure(figsize=(8,4))
sns.barplot(x=top_hosts.index.astype(str), y=top_hosts.values)
plt.title("Top Hosts by Number of Listings")
plt.xlabel("Host ID")
plt.ylabel("Listings Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# price dependency on neighbourhood
sns.barplot(data = new_df, x = 'neighbourhood_group', y = 'price', hue = 'room_type')
plt.title("Price dependency on neighbourhood")
plt.grid()
plt.show()
```

Insight:
Manhattan shows the highest average prices for private rooms, entire homes/apartments, and hotel rooms, indicating it is the most expensive neighbourhood group for Airbnb listings. Across all neighbourhoods. Shared rooms appear to have relatively higher prices in Staten Island compared to other areas.

```python
# Number of reviews and price 

sns.scatterplot(data = new_df, x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
plt.title("Price dependency on number of reviews and neighbourhood")
plt.show()
```

Insight:
Listings with higher prices receive fewer reviews compared to lower-priced listings.
